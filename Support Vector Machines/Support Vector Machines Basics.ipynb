# Support Vector Machines Basics 

## What is SVM?

Support Vector Machines (SVMs) are supervised learning models with associated learning algorithms that analyse data and recognize patterns, used for classification and regression analysis

Given a set of training examples, each marked for belonging to one of two categories (binary classification), an SVM training algorithm builds a model that assigns new examples into one category or the other, making it a non-probablistic binary linear classifier

## SVM Example

Imagine the training data below. We have two classes - Blue class and Pink class. We want to do binary classification for new data points. This means that we want to classify whether the new data point belong to the pink class or blue class. 

We can do this by drawing a separate "hyperplane" between the classes. In 2D, it is just a line between the classes. Now there can be many hyperplanes (lines) that can separate the two classes (In the below example, pink, black and green). How do we decide which line separates the classes the best?

The solution to our problem is to choose the hyperplane that maximizes the margin between classes. In the below diagram, the dotted hyperplane looks like the best choice as it has the maximum margin (the solid lines) between both the classes

The vector points (training points) that the margin line touch are known as the Support Vectors

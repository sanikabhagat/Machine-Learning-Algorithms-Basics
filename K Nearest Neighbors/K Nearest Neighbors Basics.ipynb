# K Nearest Neighbors

## What is K Nearest Neighbors?

In simple words, K Nearest Neighbors is a classification algorithm that assumes similar things exist in close proximity. In other words, similar things are near to each other

## How does KNN Work?

### Training Algorithm
1. Store all the data


### Prediction Algorithm
1. Calculate the distance from x (the new data point whose class is to be predicted) to all the points in your data
2. Sort the points in your data by increasing distance from x
3. Predict the majority label of the "k" closest points

## Explanation of KNN

Consider the below figure. Yellow points belong to Class A and purple points belong to Class B. The red start is a new data point which we have to predict which class it belongs to. Now, the value of K chosen affects what class the new point is assigned to. 

For example, if the value of k is 3, then the three nearest points to the new point are chosen. Then the three points are evaluated as to class the data points belong. Whichever class has majority, the new data point belongs to that class.

Now in the below example, for k=3 the new data point will belong to class B (as there are two purple points as comapred to one yellow point). Whereas if k=6, then the new data point will belong to class A (as there are 4 yellow points as compared to two purple points)

### Advantages of KNN:

1. Very Simple
2. Training is trivial
3. Works with any number of classes
4. Easy to add more data
5. Few parameters (There are just two parameters: "K" which is how many points are you going to look at to predict the class of your test point and "Distance Metric" which is distance between the test point and all training points) 

### Disadvantages of KNN:

1. High Prediction Cost (Worse for large dataset because you sort the data points to test what data points are closr to the test data point)
2. Not good with high dimensional data (As your features increase and you get high dimensional data, it is going to throw off your ability to measure distances in various dimensions)
3. Categorical features do not work well
